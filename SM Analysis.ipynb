```python
import pandas as pd
sm=pd.read_excel("popu_soci.xlsx")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the merged_data dataframe by Percentages in descending order and select the top 10 rows
top_10_countries = sm.nlargest(10, 'Percentages')

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='Percentages', y='Country', data=top_10_countries, palette='viridis')

# Adding population and social media user numbers as text annotations
for i, (percentage, country, population, social_media_users) in enumerate(zip(top_10_countries['Percentages'], 
                                                                              top_10_countries['Country'],
                                                                              top_10_countries['Population'],
                                                                              top_10_countries['SocialMediaUsersTotal2023'])):
    plt.text(percentage + 2, i, f'Population: {population}\nSocial Media Users: {social_media_users}', va='center')

# Adding labels and title
plt.xlabel('Percentage of Social Media Users')
plt.ylabel('Country')
plt.title('Top 10 Countries with Highest Percentage of Social Media Users (2023)')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
