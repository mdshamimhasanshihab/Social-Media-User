```python
import numpy as np
data = {
    'Platform': ['Facebook', 'Instagram', 'Snapchat', 'Twitter', 'Tik Tok', 'Youtube', 'LinkedIn'],
    'Male Users': [56.6, 50.7, 45.4, 56.4, 43, 53.9, 57.2],
    'Female Users': [43.4, 49.3, 53.8, 43.6, 57, 46.1, 42.8]
}

data = {
    'Platform': ['Facebook', 'Instagram', 'Snapchat', 'Twitter', 'Tik Tok', 'Youtube', 'LinkedIn'],
    'Male Users': [56.6, 50.7, 45.4, 56.4, 43, 53.9, 57.2],
    'Female Users': [43.4, 49.3, 53.8, 43.6, 57, 46.1, 42.8]
}

# Set color scale
colors = plt.cm.viridis(np.linspace(0, 1, len(data['Platform'])))

# Bar width
bar_width = 0.4

# Bar positions
bar_positions_male = np.arange(len(data['Platform']))
bar_positions_female = bar_positions_male + bar_width

# Create the figure and axis with a size of 16x12
fig, ax = plt.subplots(figsize=(16, 12))

# Plot bars for Male Users
ax.bar(bar_positions_male, data['Male Users'], width=bar_width, color=colors, label='Male Users', edgecolor='black')

# Plot bars for Female Users
ax.bar(bar_positions_female, data['Female Users'], width=bar_width, color=colors, label='Female Users', edgecolor='black', alpha=0.7)

# Set x-axis ticks and labels
ax.set_xticks(bar_positions_male + bar_width / 2)
ax.set_xticklabels(data['Platform'])

# Display values on each bar
for i, (male, female) in enumerate(zip(data['Male Users'], data['Female Users'])):
    ax.text(bar_positions_male[i] + bar_width / 2, male + 1, f"{male:.1f}%", ha='center', va='bottom')
    ax.text(bar_positions_female[i] + bar_width / 2, female + 1, f"{female:.1f}%", ha='center', va='bottom')

# Set labels and title
ax.set_xlabel('Social Media Platform')
ax.set_ylabel('Percentage of Users')
ax.set_title('Male and Female Users on Social Media Platforms')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python


# Read the Excel file
file_path = "Social Media User.xlsx"
df = pd.read_excel(file_path)

# Function to plot the top 10 countries for a specific social media platform
def plot_top_countries(platform_column, title):
    top_countries = df.nlargest(10, platform_column)[['Country', platform_column]]

    # Use Seaborn for a modern look and add a color scale
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Country', y=platform_column, data=top_countries, palette='viridis')
    plt.xlabel('Country')
    plt.ylabel(f'{title} Users (In million)')
    plt.title(f'Top 10 Countries with the Highest {title} Users')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
    plt.tight_layout()

    # Show the plot
    plt.show()

# List of social media platforms and corresponding columns in the DataFrame
platforms = [
    ('Facebook', 'Facebook Users'),
    ('Instagram', 'Instagram Users'),
    ('Snapchat', 'Snapchat Users'),
    ('TikTok', 'TikTok Users'),
    ('Twitter', 'Twitter Users'),
    ('WhatsApp', 'WhatsApp Users'),
    ('LinkedIn', 'LinkedIn Users')
]

# Set Seaborn style
sns.set_theme()

# Plot the top 10 countries for each social media platform
for platform, column in platforms:
    plot_top_countries(column, platform)
```
