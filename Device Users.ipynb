```python
Total_population=169828921
female_popu=85653120
male_popu=84077203
other_popu=Total_population-(female_popu+male_popu)
```

```python
laptop_user=(female_popu*1.7+male_popu*4.8)/100
desktop_user=(female_popu*1.6+male_popu*3.4)/100
tablet_user=(female_popu*1.9+male_popu*2.1)/100
```

```python
categories = ['Laptop Users', 'Desktop Users', 'Tablet Users']
users = [laptop_user, desktop_user, tablet_user]

plt.bar(categories, users, color=['blue', 'green', 'orange'])

# Add labels and title
plt.xlabel('Device Types')
plt.ylabel('Number of Users')
plt.title('Device Users in the Population')

# Display the actual values on the bars
for i, user in enumerate(users):
    plt.text(i, user, f'{user:.0f}', ha='center', va='bottom')

# Show the plot
plt.show()
```
